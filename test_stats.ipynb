#  Intervalles de confiance

## Variables qualitatives

Une variable *qualitative* est un type de variable qui mesure une qualité d'un individu statistique. Dans la suite, on s'intéressera uniquement à des variables qualitatives binaires:
- garçon/fille,
- a voté/ n'a pas voté,
- fort en physique/nul en physique.  




Il est possible de "qualitativer" une variable quantitative en créant des catégories:






| variable quantitative | équivalent qualitatif |
| ------ | ------ |
| poids | poids < 90 kg ou poids > 90 kg |
| score de fatigue | plus fatigué ou moins fatigué |


On définit la *proportion* (ou *prévalence*) d'une certaine qualité (par ex: avoir un poids supérieur à 90 kg) comme le nombre d'individus qui ont cette qualité sur le nombre total d'individus.


Lorsque la taille de l'échantillon *n* est grande, on a vu qu'un intervalle de confiance à 95% de la prévalence *p* dans la population générale est donnée par:

$$ p_{ech} - 1.96\sqrt{p_{ech}(1-p_{ech})\over n} \leq p \leq  p_{ech} + 1.96\sqrt{p_{ech}(1-p_{ech})\over n} $$

où $p_{ech}$ est la prévalence de la qualité dans l'échantillon.




Lorsque *n* n'est pas suffisament grand, cette formule n'est plus valable car on ne peut plus approximer la loi binomiale par une loi normale.

## Variables quantitatives

Une variable *quantitative* est un type de variable qui mesure une propriété *métrique* d'un individu statistique, elle s'exprime à l’aide de nombres, et ses diverses valeurs peuvent être numériquement comparées.

Des exemples sont:
- le temps de course,
- le poids,
- la longueur de foulée...


Considérons la variable quantitative $T = $*temps de course sur 10 km*.

On cherche à connaître le temps moyen $\tau$  pour une population de taille $N$ mais on ne dispose que des valeurs pour un échantillon de taille $n << N$.

 Dans ce cas, on peut calculer l'intervalle de confiance à 95 % à partir de la relation:

$$ \overline{T} - \eta {S \over \sqrt{n}} \leq \tau \leq  \overline{T} + \eta {S \over \sqrt{n}}  $$

avec:

*   $\overline{T} = {1 \over n} \sum_{i = 1}^n T_i$, moyenne du temps sur l'échantillon.
*   $S = \sqrt{\sum_{i=1}^n (\overline{T} - T_i)^2 \over n-1}$, écart-type non biaisé de l'échantillon,
*   $\eta$: quantile à 97,5% de la loi de Student à $n-1$ degrés de liberté.

Cette dernière grandeur est contenue dans des tables, comme pour la loi normale. On peut facilement l'évaluer avec la fonction *stats.t.ppf* sous python.


```python
import scipy.stats as stats
from math import sqrt

#calcul de la moyenne, de l'écart type non biaisé et de la taille de l'échantillon

mean = runners['temps_10km'].mean()
S = runners['temps_10km'].std()
n = runners['temps_10km'].count()

#calcul du quantile à 97,5%

eta = stats.t.ppf(q = 0.975, df=n-1)

marge= eta*S/sqrt(n)

intervalle_conf = (mean - marge, mean + marge)
print(intervalle_conf)
```

# Test d'hypothèse: le test de Student apparié

Le **test de Student apparié** est une méthode utilisée pour tester si la différence moyenne entre des paires de mesures est nulle ou pas.
Par paires de mesures, on entend des mesures effectuées sur un même groupe de personne à des temps différents.  



Prenons l'exemple d'un enseignant qui souhaite savoir si deux examens sur le même sujet sont aussi difficiles l'un que l'autre.

Pour cela, il donne à des étudiants le premier examen et le jour suivant, il donne aux étudiants le deuxième examen. Ensuite, il corrige.

A la fin de sa correction, il dispose de 2$n$ notes, où $n = 20$ est le nombre d'étudiants.



```python
# Set random seed for reproducibility
import numpy as np
np.random.seed(34)

# On crée ici deux faux jeux de données pour les notes aux examens. On utilise
# pour cela deux jeux de données suivant une loi gaussienne dont les moyennes et
# écart-type ont été choisis arbitrairement.

# Notes aux premier examen
size1 = 20
mean1 = 11
std1 = 1.6
data1 = np.random.normal(mean1, std1, size1)

# Notes au deuxième examen
size2 = 20
mean2 = 9.7
std2 = 2
data2 = np.random.normal(mean2, std2, size2)

# Création de l'histogramme
plt.figure(figsize=(10, 6))
plt.hist(data1, bins=20, alpha=0.5, label=f'Examen 1 (μ={mean1}, σ={std1})')
plt.hist(data2, bins=20, alpha=0.5, label=f'Examen 2 (μ={mean2}, σ={std2})')

#plt.title('Données fictives pour deux distributions de taille 10000')
plt.xlabel('Notes ')
plt.ylabel('Fréquence')
plt.legend(loc='upper right')
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

D'après la valeur des deux moyennes et l'allure des histogrammes, le premier examen a été mieux réussi que le second. Peut-on en conclure toutefois que le premier examen est *objectivement* plus facile?


Pour s'en assurer, l'enseignant va mener un test de Student apparié en supposant que les deux sujets sont de même difficulté (hypothèse $H_0$).

En calculant la valeur $p$, on aura une estimation de la probabilité que cet effet apparemment positif soit dû au hasard. On quantifie ainsi la probabilité d'être tombé sur un **faux positif**.

En Python, il suffit d'utiliser la fonction *ttest_rel*.

```python
stats.ttest_rel(data1,data2)
```

Le test renvoie la valeur $p = 0.18$. Si on prend $\alpha = 0.05$ comme cela est souvent fait dans la littérature scientifique, on peut dire le test n'est pas concluant. On n'a pas assez d'élément pour rejeter $H_0$.

Pris d'un doute, l'enseignant décide de refaire le test mais avec 40 étudiants. La moyenne et l'écart-type sont identiques pour les deux tests:

```python
# Notes aux premier examen
size3 = 40
mean3 = 11
std3 = 1.6
data3 = np.random.normal(mean3, std3, size3)

# Notes au deuxième examen
size4 = 40
mean4 = 9.7
std4 = 2
data4 = np.random.normal(mean4, std4, size4)

stats.ttest_rel(data3,
                data4)
```

Le test renvoie $p=8.3\times 10^{-6}$. Cette fois, pas de doute, il est hautement improbable que les examens aient les mêmes degrés de difficulté.

## Remarques



1.   Un test de Student apparié ne fonctionne que pour des données ayant à peu près une distribution normale. Plus la distribution des notes s'éloigne de la normale, moins le test est fiable.

2.   Comme on le voit sur l'exemple ci-dessus, le résultat du test dépend fortement de la taille de l'échantillon.



## Puissance d'un test

Imaginons maintenant une situation différente où on sait que deux examens n'ont pas la même difficulté. On fait passer chaque examen à un échantillon de 20 étudiants et on observe une différence de moyenne très faible.  Il y a alors un risque d'accepter l'hypothèse $H_0$ alors que $H_1$ est vrai. C'est ce qu'on appelle un **faux négatif**.


La *puissance* d'un test désigne la capacité d'un test à trouver un effet positif dans l'échantillon alors que l'effet est effectivement positif pour toute la population. Elle vaut $P = 1 - \beta$, où $\beta$ est le risque de deuxième espèce.

En général, un test est de qualité si la puissance vaut au moins 0.8.

Il y a deux problèmes concernant la puissance:

1.   il est difficile de l'estimer directement car on n'a pas assez d'informations sur $H_1$  (par exemple, on ne connaît pas la différence de moyenne des notes pour toute la population).
2.   lorsqu'on choisit une valeur de $\alpha$ pour un test, il y a une balance entre la valeur de $\alpha$ et celle de $\beta$. Si on prend un $\alpha$ très petit, comme 0.01, le test aura peu de faux positifs mais possiblement plus de faux négatifs.

La seule solution pour diminuer à la fois $\alpha$ et $\beta$ est d'augmenter la taille de l'échantillon.






Il est possible de déterminer la taille de l'échantillon pour avoir un $\alpha$ et un $\beta$ donné à l'aide de la fonction *statsmodels.stats.power.tt_ind_solve_power*.

```python
from statsmodels.stats.power import tt_solve_power

tt_solve_power(effect_size = 0.5,
               alpha = 0.05,
               power = 0.8)
```

Pour avoir un risque $\alpha = 0.05$ et un test de puissance 0.8, la taille de l'échantillon doit vérifier $n \geq 34$.

# Régression linéaire: test de nullité du coefficient directeur

 Considérons le nuage de points représentant la fréquence cardiaque après effort avec le poids du coureur et déterminons la droite de régression linéaire $y = ax+ b$.

```python
#nuage de points
g = sb.scatterplot(x="poids", y="fc_effort", data=runners)

# Initialize model
regression_model = linear_model.LinearRegression()

# Mise en forme des données
regression_model.fit(X=runners['poids'].values.reshape(-1,1),
                     y=runners['fc_effort'].values.reshape(-1,1))

prediction_lineaire = runners['poids'].values.reshape(-1,1)*regression_model.coef_+regression_model.intercept_
plt.plot(runners["poids"].values.reshape(-1,1),
         prediction_lineaire,
         color="blue", label=f"a={regression_model.coef_[0]}, b={regression_model.intercept_}");

## ajout de titres et d'axes
plt.title("Fréquence cardiaque après effort en fonction du poids")
plt.legend()
plt.xlabel('Poids (kg) ')
plt.ylabel('Fréquence cardiaque (bpm)');
```

On voit que le nuage de points est très éparpillé et que la droite de régression linéaire est quasiment horizontal. Néanmoins, le coefficient directeur n'est pas tout à fait nul et vaut $a = 0.276  >0$. Pour notre échantillon, plus le coureur est lourd, plus la fréquence cardiaque après effort est élevée. Peut-on néanmoins conclure que cela est vrai pour toute la population? Ou cela résulte-t-il du hasard lors du choix de l'échantillon?

Pour le savoir, on peut de nouveau mener un test statistique.

Notons $H_0 = \{ a = 0 \}$ et $H_1 = \{ a \neq 0 \}$. On suppose de nouveau l'hypothèse nulle et on calcule la valeur $p$. Si $p < \alpha$, on pourra rejeter l'hypothèse nulle.

Pour ce faire, on utilisera la librairie *statsmodels.api*

```python
import statsmodels.api as sm

x = runners['poids']
y = runners['fc_effort']

#on rajoute un bruit gaussien pour tester l'hypothèse.
x = sm.add_constant(x)

#On calcule de nouveau la régression linéaire
model = sm.OLS(y, x).fit()

#On imprime le résumé de l'étude statistique.
print(model.summary())
```

La valeur $p$ correspondant au coefficient directeur est la valeur écrite après en face de l'expression "Prob (F-statistic)".

Ici, on trouve $p = 0.271$. On ne peut donc pas rejetter $H_0$.
